```python
x = arange(0,30)
```

```python
x = x.reshape((30,1))
```

```python
error = np.random.normal(0,15, (30,1))
```

```python
y = 5*x+2+error
```

```python
scatter(x,y)
```

```python
A= hstack([x, ones_like(x)])
```

```python
v = linalg.lstsq(A,y)[0]
```

```python
plot(x, dot(A, v), "r")
scatter(x,y)
```
